# How Loop Executes

* Counter/Flag ----> Condition ----> Increment
* First Loop was While loop

```python
counter = 0
while counter <= 11:
    print(counter)
    
    counter+=1
```

```python
counter = 100
while counter >= 51:
    print (counter)
    counter-=1
```

```python
y = 2023
while y > 1999:
    print(y)
    y-=1
```

```python
n = ["Sha", "Piy", "Jkl"]

i = 0
while i < 2:
    print(n[i])
    i+=1
```

# To print all the values in the list without knowing the indexing

```python
n = ["Sha", "Piy", "Jkl"]

i = 0
while i < len(n):
    print(n[i])
    i+=1
```

# Drawbacks of while loop

* Manually setting the counter
* Logic tests on Boolean (True/False)
* Increment or decrement the counter

# The solution is for loop

# For Loop

```python
n = ["Sha", "Piy", "Jkl"]
for b in n:
    print(b)
```

# Enumerate returns the index no. and returns the tupple

```python
n = ["Sha", "Piy", "Jkl"]
a = list(enumerate(n))
print(a)
list(enumerate(n))
```

```python
n = ["Sha", "Piy", "Jkl"]
for idx, value in enumerate(n):
    print(f"Index: {idx} and Value = {value}")
```

```python
n = ["Sha", "Piy", "Jkl"]
for idx, value in enumerate(sorted(n)):
    print(f"Index: {idx} and Value = {value}")
```

# Dynamic System ----> Run until user wish to work

```python
i = 1
while i < 1000:
    if (i % 5==0) and (i % 7 ==0) and (i % 20 ==0):
        print(i)
    i+=1
```

# Break means we can come out of the loop while reaching to dedicated condition


```python
i = 0
while i < 7:
    if i ==5:
        break 
    print (i)
    i+=1
    
```

```python
i = 0
while i < 7:
    print (i)
    if i ==5:
        continue
    i+=1
```

# Continue will not execute the block, command etc when the condition become true

```python
for i in range(11):
    if i == 5:
        continue
    print(i)
```

# Pass---> It is to pass the block command to ignore this time because it will be written later. 
## This will not generate error

```python
for i in range(11):
    pass 
```
